```python
import pandas as pd
import bs4
import requests as r
import unicodedata
import pandas as pd
import glob
import numpy as np
```

```python
def seasonscrape(season):
    url = f"https://www.basketball-reference.com/leagues/NBA_{season}_totals.html"
    req = r.get(url)
    soup = bs4.BeautifulSoup(req.text, 'lxml')
    header = list()

    for i in soup.find('div',dict(id='div_totals_stats')).find('thead').find_all('th'):
        header.append(i.text)

    header = header[1:]

    player_table=[]
    for i in soup.find('div',dict(id='div_totals_stats')).find('tbody').find_all('tr'):
        player_row=[]
        for j in i.find_all('td'):
            player_row.append(j.text)
        player_table.append(player_row)
        
    
    df = pd.DataFrame(player_table, columns = header)
    df['Season'] = season
    df['unique'] = df['Player']+'_'+df['Tm']+'_'+(str(season))
    df.set_index('unique')
    return df
```
