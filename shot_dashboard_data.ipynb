```python
def shot_dash(season):
    import pandas as pd
    import requests
    import time
    
    headers = {
'Host': 'stats.nba.com',
'Connection': 'keep-alive',
'Accept': 'application/json, text/plain, */*',
'x-nba-stats-token': 'true',
'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36',
'x-nba-stats-origin': 'stats',
'Referer': 'https://stats.nba.com/players/shots-closest-defender/',
'Accept-Encoding': 'gzip, deflate, br',
'Accept-Language': 'en-US,en;q=0.9'}

#Closest Defender
    distance = ['0-2+Feet+-+Very+Tight','2-4+Feet+-+Tight','4-6+Feet+-+Open','6%2B+Feet+-+Wide+Open']
    for defenderdistance in distance:
        url = f"https://stats.nba.com/stats/leaguedashplayerptshot?CloseDefDistRange={defenderdistance}&College=&Conference=&Country=&DateFrom=&DateTo=&Division=&DraftPick=&DraftYear=&DribbleRange=&GameScope=&GameSegment=&GeneralRange=&Height=&LastNGames=0&LeagueID=00&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season={season}&SeasonSegment=&SeasonType=Regular+Season&ShotClockRange=&ShotDistRange=&StarterBench=&TeamID=0&TouchTimeRange=&VsConference=&VsDivision=&Weight="
        r = requests.get(url, headers=headers).json()

        df = pd.DataFrame(r['resultSets'][0]['rowSet'], columns = r['resultSets'][0]['headers'])
        df_title = f"{season}_shotdash_{defenderdistance}.csv"
        df.to_csv(df_title, index=False)
        time.sleep(10)
        
#General        
    headers.update(Referer=f"https://stats.nba.com/players/shots-general/?Season={season}&SeasonType=Regular%20Season")
    where = ['Overall','Catch+and+Shoot','Pullups','Less+Than+10+ft']
    for shotrange in where:
        url = f"https://stats.nba.com/stats/leaguedashplayerptshot?CloseDefDistRange=&College=&Conference=&Country=&DateFrom=&DateTo=&Division=&DraftPick=&DraftYear=&DribbleRange=&GameScope=&GameSegment=&GeneralRange={shotrange}&Height=&LastNGames=0&LeagueID=00&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season={season}&SeasonSegment=&SeasonType=Regular+Season&ShotClockRange=&ShotDistRange=&StarterBench=&TeamID=0&TouchTimeRange=&VsConference=&VsDivision=&Weight="
        r = requests.get(url, headers=headers).json()

        df = pd.DataFrame(r['resultSets'][0]['rowSet'], columns = r['resultSets'][0]['headers'])
        df_title = f"{season}_shotdash_{shotrange}.csv"
        df.to_csv(df_title, index=False)
        time.sleep(10)

#Shot Clock
    headers.update(Referer='https://stats.nba.com/players/shots-shotclock/')
    seconds = ['24-22','22-18+Very+Early','18-15+Early','15-7+Average','7-4+Late','4-0+Very+Late','ShotClock+Off']
    for clock in seconds:
        url = f"https://stats.nba.com/stats/leaguedashplayerptshot?CloseDefDistRange=&College=&Conference=&Country=&DateFrom=&DateTo=&Division=&DraftPick=&DraftYear=&DribbleRange=&GameScope=&GameSegment=&GeneralRange=&Height=&LastNGames=0&LeagueID=00&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season={season}&SeasonSegment=&SeasonType=Regular+Season&ShotClockRange={clock}&ShotDistRange=&StarterBench=&TeamID=0&TouchTimeRange=&VsConference=&VsDivision=&Weight="
        r = requests.get(url, headers=headers).json()

        df = pd.DataFrame(r['resultSets'][0]['rowSet'], columns = r['resultSets'][0]['headers'])
        df_title = f"{season}_shotdash_{clock}.csv"
        df.to_csv(df_title, index=False)
        time.sleep(10)

#Dribbles
    headers.update(Referer='https://stats.nba.com/players/shots-dribbles/')
    dribbles = ['0+Dribbles','1+Dribble','2+Dribbles','3-6+Dribbles','7%2B+Dribbles']
    for drib in dribbles:
        url = f"https://stats.nba.com/stats/leaguedashplayerptshot?CloseDefDistRange=&College=&Conference=&Country=&DateFrom=&DateTo=&Division=&DraftPick=&DraftYear=&DribbleRange={drib}&GameScope=&GameSegment=&GeneralRange=&Height=&LastNGames=0&LeagueID=00&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season={season}&SeasonSegment=&SeasonType=Regular+Season&ShotClockRange=&ShotDistRange=&StarterBench=&TeamID=0&TouchTimeRange=&VsConference=&VsDivision=&Weight="
        r = requests.get(url, headers=headers).json()

        df = pd.DataFrame(r['resultSets'][0]['rowSet'], columns = r['resultSets'][0]['headers'])
        df_title = f"{season}_shotdash_{drib}.csv"
        df.to_csv(df_title, index=False)
        time.sleep(10)
        
#Touch Time
    headers.update(Referer='https://stats.nba.com/players/shots-touch-time/')
    possession = ['Touch+%3C+2+Seconds','Touch+2-6+Seconds','Touch+6%2B+Seconds']
    for touch in possession:
        url = f"https://stats.nba.com/stats/leaguedashplayerptshot?CloseDefDistRange=&College=&Conference=&Country=&DateFrom=&DateTo=&Division=&DraftPick=&DraftYear=&DribbleRange=&GameScope=&GameSegment=&GeneralRange=&Height=&LastNGames=0&LeagueID=00&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season={season}&SeasonSegment=&SeasonType=Regular+Season&ShotClockRange=&ShotDistRange=&StarterBench=&TeamID=0&TouchTimeRange={touch}&VsConference=&VsDivision=&Weight="
        r = requests.get(url, headers=headers).json()

        df = pd.DataFrame(r['resultSets'][0]['rowSet'], columns = r['resultSets'][0]['headers'])
        df_title = f"{season}_shotdash_{touch}.csv"
        df.to_csv(df_title, index=False)
        time.sleep(10)
        
#Closest Defender on Shots 10+ Feet Away
    headers.update(Referer='https://stats.nba.com/players/shots-closest-defender-10/')
    dist = ['0-2+Feet+-+Very+Tight','2-4+Feet+-+Tight','4-6+Feet+-+Open','6%2B+Feet+-+Wide+Open']
    for jumperdist in dist:
        url = f"https://stats.nba.com/stats/leaguedashplayerptshot?CloseDefDistRange={jumperdist}&College=&Conference=&Country=&DateFrom=&DateTo=&Division=&DraftPick=&DraftYear=&DribbleRange=&GameScope=&GameSegment=&GeneralRange=&Height=&LastNGames=0&LeagueID=00&Location=&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season={season}&SeasonSegment=&SeasonType=Regular+Season&ShotClockRange=&ShotDistRange=%3E%3D10.0&StarterBench=&TeamID=0&TouchTimeRange=&VsConference=&VsDivision=&Weight="
        r = requests.get(url, headers=headers).json()

        df = pd.DataFrame(r['resultSets'][0]['rowSet'], columns = r['resultSets'][0]['headers'])
        df_title = f"{season}_shotdash_{jumperdist}_jumpers10ft.csv"
        df.to_csv(df_title, index=False)
        time.sleep(10)
```

```python
def allseasons():
    for choose_season in ['2019-20','2018-19','2017-18','2016-17','2015-16','2014-15','2013-14']:
        shot_dash(choose_season)
```
