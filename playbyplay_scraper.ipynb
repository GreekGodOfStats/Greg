```python
def get_pbp_detail(season,game):
    url = f"https://data.nba.com/data/10s/v2015/json/mobile_teams/nba/{season}/scores/pbp/{game}_full_pbp.json"
    r = requests.get(url).json()
    
    df_list = []
    for i in r['g']['pd']:
        df = pd.DataFrame(i['pla']) # captures the play-by-play data
        df.insert(0, 'period', i['p']) # grabs the period 
        df.insert(0, 'game',game)
        df_list.append(df)
        
    return pd.concat(df_list)

frame = get_pbp_detail('2017','0021700001')
len(frame)
```

```python
def build_game_list(season):
    gamelist=[]
    year = season[2:]
    start = int(f"{year}00001")
    end = int(f"{year}01261")
    gamenums = [number for number in range(start,end)]
    for number in gamenums:
        game = '002'+(str(number))
        gamelist.append(game)
    return gamelist
```

```python
build_game_list('2017')
```

```python
def combine_pbp_records(season): 
    result_list = build_game_list(season)
    for game in result_list:
        time.sleep(1.7)
        try:
            result = get_pbp_detail(season,game)
            result_list.append(result)
            df = pd.DataFrame(result)
            df.to_csv(f"C:\\Users\gsteele\PlaybyPlay\{game}_PlaybyPlay.csv", index=False, encoding='utf-8-sig')
        except:
            LookupError
        continue
    
    
    return result_list
```
